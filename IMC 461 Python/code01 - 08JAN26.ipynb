```python
#assign a data frame called irisdata
irisdata = pd.read_csv('datasets/irisdata.csv')

#use the .info() function to display the fields and structure within a defined data frame
irisdata.info()

#display the top n rows
irisdata.head(10)
```

```python
#display all columns in irisdata, but limit number of rows to 5
print(irisdata.head(5))

# to know the number of instances in the table
# to count number of rows in the table, count(*) function of sql is used, 
# it will display number of rows in the table
print(len(irisdata))
```

```python
# .loc conditionals
#.loc is by default .loc[:,:], with : meaning get all (same as * in SQL), first number is rows, and second is columns
#when you have only a conditional in .loc, it will refer to row conditionals, so it will filter out to get rows while still
#keeping all the columns

# display all the data related to iris flowers that are of virginica species
irisdata.loc[irisdata.species == 'virginica']

#display all the data related to iris flowers that have sepal length greater than 1.5
irisdata.loc[irisdata.sepal_length > 1.5]

# LOGICAL OPERATORS
# now, display all data related to iris flowers that are of virginica species and also have Sepal.Length greater than 7.0
irisdata.loc[(irisdata.species == 'virginica') & (irisdata.sepal_length > 7)]
```

```python
# Can't ignore strings
# keyword matching is a lot more complicated in Python, though, because of regex rules, so I won't be using regex
#if you're interested in learning more about it, you can reach out to me directly


# multiple characters
### display rows with Species "setosa"
irisdata[irisdata.species.str.startswith('set')]


### display rows with Species "versicolor"
irisdata[irisdata.species.str.endswith('color')]
```

```python
# I just want Petal Length and Petal Width

irisdata.loc[:, ['petal_length', 'petal_width']]

# I want to give my own names to the columns
# It's impractical to do this the same way in Python as we did in SQL, I can show you both ways, but since Python is for coding,
#we will usually just save the changed name into the dataframe directly or create a new variable to store it

#SQL way
irisdata.loc[:, ['petal_length', 'petal_width']].rename(columns = {'petal_length': 'PLeng', 'petal_width': 'PWidth'})
#Kind of dumb right

#2 other ways
#Replace in same variable
#irisdata.rename(columns = {'petal_length': 'PLeng', 'petal_width': 'PWidth'}, inplace = True)


#Make a new variable (best practice)
irisdata_renamed = irisdata.rename(columns = {'petal_length': 'PLeng', 'petal_width': 'PWidth'})
irisdata_renamed
```

```python
# I want to know unique species of iris flower
irisdata.species.unique()

# Let's sort Petal Length in ascending order
# use sort_values to sort the values in a column
irisdata.sort_values(by = 'petal_length').head(25)

# now in descending order
irisdata.sort_values(by = 'petal_length', ascending = False).head(25)

## We can also create new columns and perform calculations, a lot easier to do in Python
irisdata['ratio'] = irisdata.petal_length/irisdata.petal_width

#now sorting by both columns
irisdata.sort_values(by = ['petal_length', 'petal_width'], ascending = False).head(25)
```
